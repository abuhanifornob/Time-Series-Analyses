**Resampling**

```python
df['Close'].resample('M').mean()
```

```python
df['Close'].resample('D').mean()
```

```python
df['Close'].resample('Q').mean().plot(kind='bar')
```

```python
df['Open'].resample('M').sum().plot(kind='bar')
```

```python
df['Open'].resample('M').sum().plot()
```
